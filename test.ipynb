```python
from cjio import cityjson
```

```python
cm = cityjson.load("citydata/berlin/Charlottenburg-Wilmersdorf/citygml.json")
```

```python
buildings = cm.get_cityobjects(type="building")
```

```python
import random
```

```python
obj = random.choice(list(buildings.values())).geometry[0]
```

```python
obj.lod
```

```python
faces, vertmap, n_vertices = obj.build_index()
```

```python
faces_pyv = [[len(f[0]), *f[0]] for f in faces]
faces_pyv = [g for f in faces_pyv for g in f]
```

```python
verts = list(vertmap.keys())
```

```python
mesh = pyvista.PolyData(verts, faces_pyv)
```

```python
mesh.plot(jupyter_backend="static")
```
